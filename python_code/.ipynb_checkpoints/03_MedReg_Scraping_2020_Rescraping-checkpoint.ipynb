Wir machen kurz einen manuellen Test auf dem Server mit einigen ausgewhlten Adressen, die mit Javascript-Hinweis runtergekommen sind:



```python
import requests
import random
import time
import csv
from time import sleep

url_stamm = "https://www.medregom.admin.ch/DE/Detail/Detail?pid="
output_path = "medreg_errortest/"
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36'}

for url_id in range(30390,30409): #alle diese IDs kamen mit Javascript-Meldung zurck
    url_req = url_stamm + str(url_id) 
    try:
        r = requests.get(url_req, headers = headers, timeout=20)

        with open(output_path + "medreg_" + str(url_id) + ".html", 'w', encoding = 'utf-8') as f:
            f.write(r.text)
            f.close()
            protokoll = [[url_id, url_req, r.status_code]]
            print(protokoll)
          
        #sleep(random.uniform(2, 5))
        #im zweiten Durchgang aktiviert
   
    except: 
        r = requests.exceptions.ConnectionError
        print(url_req, 'Scraping paused')
        time.sleep(10)
        
        continue
```

``range(114941,114962)``

2. Durchgang: Holy shit, jetzt kommen sogar alle Files mit dem Javascript-Hinweis zurck. Nochmals ohne Sleep-Bremse:
3. Durchgang: wieder alle als Javascriptmeldung. Anderer Bereich?

```python
import os

filelist = os.listdir("medreg_2020_scraped_javascript/")

html = [(sub.split('_')[1]) for sub in filelist] 
url_list = [int(nr.split('.')[0]) for nr in html] 
```

```python
len(url_list)
```

```python
random.shuffle(url_list)

url_stamm = "https://www.medregom.admin.ch/DE/Detail/Detail?pid="
output_path = "medreg_rescrape2/"
headers = {'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_11_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/53.0.2785.143 Safari/537.36'}

for url_id in url_list: #alle diese IDs kamen mit Javascript-Meldung zurck
    url_req = url_stamm + str(url_id) 
    try:
        r = requests.get(url_req, headers = headers, timeout=20)

        with open(output_path + "medreg_" + str(url_id) + ".html", 'w', encoding = 'utf-8') as f:
            f.write(r.text)
            f.close()
            protokoll = [[url_id, url_req, r.status_code]]
            print(protokoll)
        
        
        with open("/Users/master/Desktop/CAS_DJ/MedReg/MedReg Scraping 2020/medreg_scraping_protokoll/rescrape2_protokoll.csv", 'a') as p:
            writer = csv.writer(p, dialect = 'excel')
            writer.writerows(protokoll)
        
        sleep(random.uniform(1, 5))
        
    except: 
        r = requests.exceptions.ConnectionError
        print(url_req, 'Scraping paused')
        time.sleep(120)
        
        continue
```
